# 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 50
```

```python
repl = counts[counts <= threshold].index
```

```python
df=pd.read_csv('dataset/customer.csv')
df['purchased'].unique()
```
