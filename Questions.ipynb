```python
from pandas import DataFrame
df1=DataFrame(
                    [
                        ["Aryan","Singh","Full-time","Administration",2,20000],
                        ["Rohan","Agarwal","Intern","Technical",3,5000],
                        ["Riya","Shah","Full-time","Administration",5,10000],
                        ["Yash","Bhatia","Part-time","Technical",7,10000],
                        ["Siddhant","Khanna","Full-time","Management",6,20000]
                        
                        
                    ]
    ,columns=["First name","Second name","Type","Department","YoE","Salary"]

                )
df1
```

Make a pivot table which shows avarage salary of each type of employee for each department

```python
import pandas as pd
table=pd.pivot_table(df1,index='Type',columns='Department',values='Salary')
table
```

Make a pivot table which shows the sum and mean of the salaries of each type of 
employee and the number of employees of each type.

```python
import pandas as pd 
tables=pd.pivot_table(df1,index='Type',values='Salary',aggfunc=['sum','mean','count'])
tables
```
